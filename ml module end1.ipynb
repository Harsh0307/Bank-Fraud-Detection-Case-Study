```python
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
x,y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
prediction
```

```python
#checking the performance of the matrix
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
confusion_matrix(y_test,prediction)
```

```python
fig, ax = plt.subplots(figsize = (8,8))
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(y_true = y_test,
                                        y_pred = prediction, ax=ax)
```

```python
print(classification_report(y_test,prediction))
```

```python
print(accuracy_score(y_test,prediction))
```

```python
#MODEL BUILDING WITH K-NN MODEL
from sklearn.neighbors import KNeighborsClassifier
knn_model=KNeighborsClassifier(n_neighbors=5)
knn_model.fit(x_train,y_train)
```

```python
knn_predictions=knn_model.predict(x_test)
```

```python
knn_predictions
```

```python
#Evaluate the performance of knn model
print(confusion_matrix(y_test,knn_predictions)) 
```

```python
fig, ax = plt.subplots(figsize = (8,8))
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(y_true = y_test,
                                        y_pred = knn_predictions, ax=ax)
```

```python
print(accuracy_score(y_test,knn_predictions))
```

```python
#Feature Scaling 
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train = sc.fit_transform(x_train) # Fitting and transforming the training data to scale it between -1 and +1
x_test=sc.fit_transform(x_test) ## Target variable for testing set
```

```python
#KNN Model
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2) #p value we have to provide as it is in minkowski formula
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
y_pred
```

```python
print(confusion_matrix(y_pred,y_test))
```

```python
fig, ax = plt.subplots(figsize = (8,8))
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(y_true = y_test,
                                        y_pred = y_pred, ax=ax)
```

```python
print(classification_report(y_pred,y_test))
```

```python
print(accuracy_score(y_pred,y_test))
```
